```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem.Draw import IPythonConsole
from rdkit.Chem import AllChem
from rdkit.Chem.Scaffolds.MurckoScaffold import GetScaffoldForMol
from rdkit.Chem.Scaffolds.MurckoScaffold import MurckoScaffoldSmiles
from rdkit.Chem.Scaffolds.MurckoScaffold import MakeScaffoldGeneric

#descripts
from rdkit.Chem import Lipinski
from rdkit.Chem import rdMolDescriptors
from rdkit.Chem import Descriptors

from scipy.spatial import distance
from sklearn import manifold
from sklearn import decomposition
from sklearn import preprocessing

# plots
from matplotlib import pyplot
import seaborn
```

```python
def open_data(link, source):
    df = pd.read_csv(link, sep=';')
    df = df.drop(['Molecule Name','Molecule Max Phase', '#RO5 Violations','Compound Key', 'Data Validity Comment', 'Target Name', 'Target Organism', 'Target Type', 'Document ChEMBL ID', 'Source ID',  'Assay Subcellular Fraction', 'Assay Parameters', 'Assay Variant Accession', 'Assay Variant Mutation', 'Target ChEMBL ID', 'Source Description', 'Document Journal', 'Document Year', 'Cell ChEMBL ID', 'Properties','Assay ChEMBL ID', 'Assay Description', 'Assay Type', 'BAO Format ID', 'BAO Label', 'Assay Organism','Assay Tissue ChEMBL ID','Assay Tissue Name', 'Assay Cell Type','Potential Duplicate', 'Uo Units', 'Ligand Efficiency BEI','Ligand Efficiency LE','Ligand Efficiency LLE','Ligand Efficiency SEI',], axis = 1)
    df.head()
    df = df[df['Standard Relation'] == '\'=\'']
    df['source'] = source
    df = df[df['Standard Value'] >= 500]
    df = df[df['Standard Value'] <= 1500]
    df = df.drop_duplicates()
    df1 = df[['Molecule ChEMBL ID', 'Smiles','source']]
    df1.reset_index(inplace=True)
    df1 = df1.drop(['index'], axis = 1)
    df1 = df1.dropna()
    df1['mfp'] = df1['Smiles'].apply(lambda x: AllChem.GetMorganFingerprintAsBitVect(Chem.MolFromSmiles(MurckoScaffoldSmiles(x)), 2, nBits = 1024))
    print(f"{source}: {len(df1)} velikost mnoziny")
    return df1
```

```python
estrogen_beta = open_data("data_nuclear_target/ligand_estrogen_receptor_beta_EC50.csv", 'estrogen_beta')
vitamin_D = open_data('data_nuclear_target/ligand_Vitamin_receptor_EC50.csv', 'vitamin_D')
glucocorticoid = open_data('data_nuclear_target/ligand_glucocorticoid_EC50.csv', 'glucocorticoid')
ROR_gamma = open_data('data_nuclear_target/ligand_ROR_gamma_EC50.csv', 'ROR_gamma')
peroxisome_proliferator_gamma = open_data('data_nuclear_target/ligands_peroxisome_proliferator_gamma_EC50.csv', 'peroxisome_proliferator_gamma')
peroxisome_proliferator_alpha = open_data('data_nuclear_target/ligands_peroxisome_proliferator_alpha_EC50.csv', 'peroxisome_proliferator_alpha')
estrogen_alpha = open_data('data_nuclear_target/ligands_estrogen_receptor_alpha_EC50.csv', 'estrogen_alpha')
androgen = open_data('data_nuclear_target/ligands_androgen_receptor_EC50.csv', 'androgen')
thyroid_hormon_beta = open_data('data_nuclear_target/ligands_thyroid_hormon_receptor_beta_1_EC50.csv', 'thyroid_hormon_beta')
```

```python
#peroxisome_proliferator_gamma, peroxisome_proliferator_alpha
df = estrogen_beta.append([estrogen_alpha,thyroid_hormon_beta,androgen,vitamin_D,glucocorticoid, ROR_gamma])
df.reset_index(level=0, inplace=True)
```

```python
sparse_mat = distance.pdist(list(df["mfp"]),metric="jaccard")
dist_mat = distance.squareform(sparse_mat)

clf = manifold.TSNE(n_components=2, metric="precomputed")
X_tsne = clf.fit_transform(dist_mat)

df = df.join(pd.DataFrame(X_tsne, columns=["tsne_x", "tsne_y"]))
```

```python
fig, ax = pyplot.subplots(figsize=(15,10))
p=seaborn.scatterplot(x="tsne_x",
                      y="tsne_y",
                      data=df,
                      hue="source",
                      legend="brief",
                      ax=ax,
                      alpha=1
                     )
```
